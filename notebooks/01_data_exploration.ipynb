# MVSeg — Data Exploration

Quick look at the TEE volumes and their 5-class GT labels.
Run from the repo root: `uv run jupyter lab`.

```python
import SimpleITK as sitk
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
from mvseg import CLASS_NAMES

DATA = Path('../data/raw')
case = sorted((DATA / 'images').glob('*.nrrd'))[0].stem
img = sitk.GetArrayFromImage(sitk.ReadImage(str(DATA / 'images' / f'{case}.nrrd')))
lab = sitk.GetArrayFromImage(sitk.ReadImage(str(DATA / 'labels' / f'{case}.nrrd')))
print(case, img.shape, 'labels:', np.unique(lab))
```

```python
z = img.shape[0] // 2
fig, ax = plt.subplots(1, 2, figsize=(9, 4))
ax[0].imshow(img[z], cmap='gray'); ax[0].set_title('image')
ax[1].imshow(lab[z], cmap='nipy_spectral', vmin=0, vmax=len(CLASS_NAMES)-1); ax[1].set_title('label')
for a in ax: a.axis('off')
plt.tight_layout()
```
